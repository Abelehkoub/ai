```python
tropical_count = reviews['description'].str.count('tropical').sum()
fruity_count = reviews['description'].str.count('fruity').sum()
descriptor_counts = pd.Series({'tropical': tropical_count, 'fruity': fruity_count})
descriptor_counts
```

```python
star_ratings = np.where(
    reviews['country'] == 'Canada',  # Canadian wines get 3 stars
    3,
    np.where(
        reviews['points'] >= 95,  # Points ≥ 95 get 3 stars
        3,
        np.where(
            reviews['points'] >= 85,  # Points ≥ 85 and < 95 get 2 stars
            2,
            1  # Points < 85 get 1 star
        )
    )
)
star_ratings
```
